```
keys= [e.split("$")[0].strip().replace("(","").replace(")","").split(",")  for e in di]
defs= [e.split("$")[1]  for e in di]
```

```
defs[20:40]
```

```
keys2=[]
for k in keys:
  k2=[]
  for e in k:
   if "~" in e:
     e2=e.split("~")
     k2=k2+e2
   else:
     k2.append(e)
  keys2.append(k2)
```

```
assert len(keys)==len(keys2)
```

```
list(map(lambda x: x.strip(), ['agaçay ', ' akaçay']))
```

```
keys3=[list(set(list(map(lambda x: x.strip(), k)))) for k in keys2]
```

```
for k in keys3:
  if "+" in " ".join(k):
    print(k)
```

```
for k in keys3:
  if len(k)> len(set(k)):
    print(k)
```

```
for i,k1 in enumerate(keys3):
  for j,k2 in enumerate(keys3):
    if i!= j:
      if ( len(k1+k2) > len(set(k1+k2))):
        print(i,k1,j,k2)
```
